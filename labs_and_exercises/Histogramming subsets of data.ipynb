```python
import matplotlib.pylab as plt
```

```python
import numpy as np
```

```python
x = np.random.random(100)
plt.hist(x,bins=100)
plt.show()
```

```python
plt.hist(x,bins=10)
plt.show()
```

```python
x = np.random.random(100000)
plt.hist(x,bins=100,histtype='stepfilled')
plt.show()
```

```python
x = np.random.normal(loc=5,scale=3.0,size=100000)
plt.hist(x,bins=50,range=(-6,16))
plt.show()
```
